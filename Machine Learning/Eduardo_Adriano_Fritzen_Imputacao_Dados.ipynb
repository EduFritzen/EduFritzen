# **Tarefa 1: Faça a inputação de atributos usando média, moda e mediana**

```python
# Coloque seu código aqui
import pandas as pd

cols_com_valores_ausentes = df.columns[df.isnull().any()].tolist()

cols_numericas = [col for col in cols_com_valores_ausentes if df[col].dtype != 'object']

df_media = df.copy()
df_moda = df.copy()
df_mediana = df.copy()

for col in cols_numericas:
    df[col] = df[col].replace('\.', '', regex=True).astype(float)
    media = df[col].mean()
    moda = df[col].mode()[0]
    mediana = df[col].median()

    df_media[col].fillna(media, inplace=True)

    df_moda[col].fillna(moda, inplace=True)

    df_mediana[col].fillna(mediana, inplace=True)

print("\nImputação do DataFrame usando média:")
print(df_media)
print("\nImputação do DataFrame usando moda:")
print(df_moda)
print("\nImputação do DataFrame usando mediana:")
print(df_mediana)
```

# **Tarefa 2: Faça a inputação de atributos explorando correlação entre variáveis**

```python
# Coloque seu código aqui
import numpy as np

df_numeric = df.select_dtypes(include=['float64', 'int64'])

correlation_matrix = df_numeric.corr()

for col in df_numeric.columns:
    for outra_col in df_numeric.columns:
        if col != outra_col:
            if pd.isnull(df.at[0, col]) and not pd.isnull(df.at[0, outra_col]):
                if abs(correlation_matrix.loc[col, outra_col]) > 0.5:
                    df.at[0, col] = np.nan_to_num(df.at[0, outra_col] * correlation_matrix.loc[col, outra_col])

print(df)
```

# **Tarefa 3: Faça a inputação de atributos explorando similaridade entre os exemplos**

```python
# Coloque seu código aqui
from sklearn.impute import KNNImputer

df_numeric = df.select_dtypes(include=['float64', 'int64'])

knn_imputer = KNNImputer(n_neighbors=5)
df_imputado = pd.DataFrame(knn_imputer.fit_transform(df_numeric), columns=df_numeric.columns)

for coluna in df.columns:
    if coluna not in df_numeric.columns:
        df_imputado[coluna] = df[coluna]

print(df_imputado)
```
